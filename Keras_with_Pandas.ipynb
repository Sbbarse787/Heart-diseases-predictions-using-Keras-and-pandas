```
import pandas as pd
df = pd.read_csv('heart.csv'    #Read the dataset using pandas
)
```

```
df.head()
```

```
df.isnull().sum()
```

```
X = df.iloc[:,0:13]       #Slicing of data set..X=[1,....13]
y = df.iloc[:,-1]          #Y=[Target variable]
```

```
from sklearn.model_selection import train_test_split     #Split the data set
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```
from sklearn.preprocessing import StandardScaler     #Standerisation of dataset
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```
pip install --upgrade keras   #install keras
```

```
import keras
from keras.models import Sequential
from keras.layers import Dense
classifier=Sequential()
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu',input_dim=13)) # add the layers in neural network..
classifier.add(keras.layers.Dropout(0.4))
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
classifier.add(keras.layers.Dropout(0.4))
classifier.add(Dense(units=1,activation='sigmoid'))
classifier.compile(optimizer='adagrad',loss='binary_crossentropy',metrics=['accuracy'])
classifier.fit(X_train,y_train,batch_size=10,epochs=100)
```

```
import numpy as np
```

```
y_pred=classifier.predict(X_test)
y_pred=(y_pred>0.5)
n=[44,1,0,120,169,0,1,144,1,2.8,0,0,1] #Add input of arrays to the model....
prediction=classifier.predict(sc.transform(np.array([n])))  #Prediction of dataset 
prediction=(prediction>0.5)
print(prediction)
```

```
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
